```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes with labels
nodes = ["Economic Conditions", "Donor Contributions", "Charity Income", "Demand for Charity Services",
         "Marketing and Communication Strategies", "Technological Innovation", "Donor Behavior Analysis",
         "Fundraising Efficiency", "Charity Sustainability"]
G.add_nodes_from(nodes)

# Add edges with directions
edges = [("Economic Conditions", "Donor Contributions"),
         ("Donor Contributions", "Charity Income"),
         ("Charity Income", "Charity Sustainability"),
         ("Demand for Charity Services", "Charity Income"),
         ("Marketing and Communication Strategies", "Donor Contributions"),
         ("Technological Innovation", "Donor Behavior Analysis"),
         ("Donor Behavior Analysis", "Marketing and Communication Strategies"),
         ("Fundraising Efficiency", "Charity Sustainability"),
         ("Charity Sustainability", "Technological Innovation")]
G.add_edges_from(edges)

# Specify the layout to spread the nodes visually appealingly
pos = nx.spring_layout(G, seed=42)

# Draw the network
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=4000, edge_color='k', linewidths=1, font_size=10, arrows=True)

# Show the plot
plt.title("Causal Loop Diagram for the Charity Sector")
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the graph
G = nx.DiGraph()

# Define nodes with their positions
nodes = {
    "Economic Conditions": (0, 2),
    "Donor Contributions": (1, 2),
    "Charity Income": (2, 2),
    "Demand for Charity Services": (2, 1),
    "Marketing and Communication Strategies": (1, 1),
    "Technological Innovation": (1, 0),
    "Donor Behavior Analysis": (2, 0),
    "Fundraising Efficiency": (3, 0),
    "Charity Sustainability": (4, 1),
    "Project Outcomes": (3, 2),
    "Project Outputs": (4, 2),
    "Partner Engagement": (0, 1)
}
G.add_nodes_from(nodes)

# Define edges and their attributes
edges = [
    ("Economic Conditions", "Donor Contributions", {'sign': '+'}),
    ("Donor Contributions", "Charity Income", {'sign': '+'}),
    ("Charity Income", "Charity Sustainability", {'sign': '+'}),
    ("Demand for Charity Services", "Charity Income", {'sign': '-'}),
    ("Marketing and Communication Strategies", "Donor Contributions", {'sign': '+'}),
    ("Technological Innovation", "Donor Behavior Analysis", {'sign': '+'}),
    ("Donor Behavior Analysis", "Fundraising Efficiency", {'sign': '+'}),
    ("Fundraising Efficiency", "Charity Sustainability", {'sign': '+'}),
    ("Project Outcomes", "Project Outputs", {'sign': '+'}),
    ("Project Outputs", "Charity Sustainability", {'sign': '+'}),
    ("Partner Engagement", "Marketing and Communication Strategies", {'sign': '+'}),
    ("Donor Contributions", "Demand for Charity Services", {'sign': '-'}),
    ("Charity Sustainability", "Economic Conditions", {'sign': '-'})
]
G.add_edges_from(edges)

# Create the positions dictionary for node placement
pos = {node: (x, y) for node, (x, y) in nodes.items()}

# Draw the nodes
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=3000)

# Draw the edges
nx.draw_networkx_edges(G, pos, edge_color='darkblue', arrows=True, arrowstyle='->')

# Label the nodes
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# Edge labels
edge_labels = nx.get_edge_attributes(G, 'sign')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, label_pos=0.5, font_size=12)

# Remove the axis and display the plot
plt.axis('off')
plt.title("Causal Loop Diagram for the Charity Sector Project", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Drawing nodes with a white fill to appear invisible against the background
nx.draw_networkx_nodes(G, pos, node_color='white', node_size=3000, edgecolors='black')

# Drawing the labels
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# Drawing edges with curved lines
for edge in edges:
    # Unpack edge info
    u, v, data = edge
    # Calculate curvature (control points for Bezier curve)
    rad = 0.2 if data['sign'] == '+' else -0.2  # Positive edges curve one way, negative the other
    nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], connectionstyle=f'arc3,rad={rad}', edge_color='black' if data['sign'] == '+' else 'red')


# Draw edge labels to indicate positive or negative influence
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, label_pos=0.5, font_size=12)

# Remove the axis and display the plot
plt.axis('off')
plt.title("Causal Loop Diagram for the Charity Sector Project", fontsize=15)
plt.show()
```

```python

custom_pos = {
    "Economic Conditions": (0, 4),
    "Donor Contributions": (1, 3),
    "Charity Income": (2, 4),
    "Demand for Charity Services": (3, 3),
    "Marketing and Communication Strategies": (1, 1),
    "Technological Innovation": (2, 0),
    "Donor Behavior Analysis": (3, 1),
    "Fundraising Efficiency": (4, 0),
    "Charity Sustainability": (5, 1),
    "Project Outcomes": (4, 2),
    "Project Outputs": (5, 3),
    "Partner Engagement": (0, 2)  # Adjusted for step-like vertical alignment
}



edges = [
    ("Economic Conditions", "Donor Contributions", {'sign': '+'}),
    ("Donor Contributions", "Charity Income", {'sign': '+'}),
    ("Charity Income", "Charity Sustainability", {'sign': '+'}),
    ("Demand for Charity Services", "Charity Income", {'sign': '-'}),
    ("Marketing and Communication Strategies", "Donor Contributions", {'sign': '+'}),
    ("Technological Innovation", "Donor Behavior Analysis", {'sign': '+'}),
    ("Donor Behavior Analysis", "Fundraising Efficiency", {'sign': '+'}),
    ("Fundraising Efficiency", "Charity Sustainability", {'sign': '+'}),
    ("Project Outcomes", "Project Outputs", {'sign': '+'}),
    ("Project Outputs", "Charity Sustainability", {'sign': '+'}),
    ("Partner Engagement", "Marketing and Communication Strategies", {'sign': '+'}),
    ("Donor Contributions", "Demand for Charity Services", {'sign': '-'}),
    ("Charity Sustainability", "Economic Conditions", {'sign': '-'})
]

# Create the graph again
G = nx.DiGraph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(15, 10))
```

```python
plt.figure(figsize=(25, 15))



# Draw curved edges with arrows
for (u, v, data) in G.edges(data=True):
    style = 'arc3,rad=0.3' if data['sign'] == '+' else 'arc3,rad=-0.3'
    color = 'black' if data['sign'] == '+' else 'red'
    nx.draw_networkx_edges(
        G,
        pos,
        edgelist=[(u, v)],
        connectionstyle=style,
        edge_color=color,
        arrowstyle='->',
        node_size=0
    )

# Draw the node labels with bold font and increased size
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

# Remove axes
plt.axis('off')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(25, 15))

# Drawing curved edges with arrows
for (u, v, data) in G.edges(data=True):
    style = 'arc3, rad=0.3' if data['sign'] == '+' else 'arc3, rad=-0.3'
    color = 'black' if data['sign'] == '+' else 'red'
    # Draw the edge with a curve
    nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], arrowstyle='->', arrowsize=20,
                           connectionstyle=style, edge_color=color)

# Draw the node labels with bold font and increased size
nx.draw_networkx_labels(G, pos, font_size=, font_weight='bold')

# Create edge labels for the signs
edge_labels = {(u, v): data['sign'] for u, v, data in G.edges(data=True)}

# Custom function to draw edge labels with larger, bold and colored signs
def draw_edge_labels_custom(G, pos, edge_labels, ax):
    for (u, v), label in edge_labels.items():
        # Get the midpoint of the edge where the label will be placed
        midpoint_x = (pos[u][0] + pos[v][0]) / 2
        midpoint_y = (pos[u][1] + pos[v][1]) / 2
        # Depending on the sign, offset the label above or below the edge
        offset = 0.1 if label == '+' else -0.1
        ax.text(midpoint_x, midpoint_y + offset, label, fontsize=16, color='blue',
                fontweight='bold', ha='center', va='center')

# Draw custom edge labels
draw_edge_labels_custom(G, pos, edge_labels, ax)

# Remove axes
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

# Draw the curved edges with arrows
for (u, v, data) in G.edges(data=True):
    # Determine the style of the edge based on the sign
    style = 'arc3, rad=0.3' if data['sign'] == '+' else 'arc3, rad=-0.3'
    # Draw the edge with a curve
    nx.draw_networkx_edges(
        G, custom_pos, edgelist=[(u, v)], arrowstyle='->', arrowsize=15,
        connectionstyle=style, edge_color='black', node_size=0
    )

# Draw the node labels with bold font and increased size
nx.draw_networkx_labels(G, custom_pos, font_size=12, font_weight='bold')

# Add edge labels for '+' and '-' signs at the midpoint of the edges
for (u, v, data) in G.edges(data=True):
    # Calculate the midpoint
    mid_x = (custom_pos[u][0] + custom_pos[v][0]) / 2
    mid_y = (custom_pos[u][1] + custom_pos[v][1]) / 2
    # Determine the sign and its color
    sign = data['sign']
    color = 'green' if sign == '+' else 'red'
    # Draw the sign on the plot
    ax.text(mid_x, mid_y, sign, fontsize=16, color=color, ha='center', va='center', fontweight='bold')

# Remove the axes
plt.axis('off')
plt.show()
```
